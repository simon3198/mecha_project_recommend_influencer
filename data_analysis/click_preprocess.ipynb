```python
import pandas as pd

data = pd.read_csv('reward_log_referer - reward_logs.csv')
data
```

```python
index = data['https://www.naver.com/'].str.contains("blog.naver.com")
# 유튜브 링크, 메카솔로션 링크 등 블로그 글이 아닌 것들을 제거하고 네이버 블로그 글만 가져옵니다.
data = data[index]
data = data.reset_index(drop=True)
data

```

```python
import re
num =-1
# 메카솔루션 자체 제작 블로그 'roboholic'이 들어간 블로그는 전처리합니다.
while num < 36600:
    num+=1
    if 'roboholic' in data['https://www.naver.com/'][num]:
        data.drop(index=num,inplace=True)

data
```

```python
data = data.reset_index(drop=True)
data
```

```python
num =-1

#링크 속 blogid 와 logno 추출 후  (blogid.logno) 형식으로 저장
while num < 20730:
    num+=1
    print(num)
    # 링크 형식이 blogid=***&logno=*** 일때
    if 'blogId' in data['https://www.naver.com/'][num]:
        try:
            found = re.search('blogId=(.+?)&', data['https://www.naver.com/'][num]).group(1)
            try:
                found += '.'
                found += re.search('logNo=(.+?)&', data['https://www.naver.com/'][num]).group(1)
            except:
                data.drop(index=num,inplace=True)
                continue
        except:
            data.drop(index=num,inplace=True)
            continue
    # 링크 형식이 www.naver.com/id/logno 일때
    else:
        # found = re.search('com/(.+?)/',data['https://www.naver.com/'][num]).group(1)
        found = data['https://www.naver.com/'][num].split('/')
        found = found[-2]+'.'+found[-1]
        print(found)
    data['https://www.naver.com/'][num] = found

data = data.reset_index(drop=True)
data
```

```python
# 블로그 글 하나당 클릭수 계산
df = pd.DataFrame(data['https://www.naver.com/'].value_counts())
df.columns.values[0] = "click"
df.to_csv('click.csv')
```
