```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = "./data/test_after_blogger.csv"
df = pd.read_csv(file_name, index_col = 0)
df.sample(5)
```

```python
import ast 
df['AP'] = df['AP'].apply(lambda x : ast.literal_eval(x))
df['AP']
```

```python
df['AP'].iloc[0]
```

```python
def sentencesAfterNooneNone(_ap):
    sentences_after_no_oneNone = 0
    for sentence, sentiment, subjective in _ap:
        if sentiment != 1 or subjective != 1:
            sentences_after_no_oneNone += 1
    return sentences_after_no_oneNone
sentencesAfterNooneNone(df.iloc[0,:]['AP']) 
```

```python
sentence_num_list = df['AP'].apply(lambda x: len(x))
sentence_num_list = sentence_num_list.values
sentence_after_no_oneNone_num_list = df['AP'].apply(sentencesAfterNooneNone)
sentence_after_no_oneNone_num_list = sentence_after_no_oneNone_num_list.values
sentence_after_no_oneNone_num_list
```

```python
X = [i+1 for i in range(90)]
```

```python
plt.plot(X, sentence_num_list, label='SentenceNum')
plt.plot(X, sentence_after_no_oneNone_num_list, label='SentenceNum(no 1/1)')
plt.xlabel('days')
plt.ylabel('sentence num')
plt.legend()
plt.show()
```

```python
plt.plot(X, sentence_num_list - sentence_after_no_oneNone_num_list, label='residual')
plt.xlabel('days')
plt.ylabel('sentence num - sentence num(no1/1)')
plt.show()
```

```python
def calculateNumOfSS1(rows, target_sent:int = 0, target_subj:int = 0):
    sentiment_sum = 0
    subjective_sum = 0
    for sentence, sentiment, subjective in rows:
        if sentiment == target_sent: sentiment_sum += 1
        if subjective == target_subj: subjective_sum += 1
    return sentiment_sum, subjective_sum
calculateNumOfSS1(df.iloc[0,:]['AP'],2,1)
```

```python
def calculateNumOfSS2(rows, target_sent:int = 0, target_subj:int = 0):
    sentiment_subjective_sum = 0
    for sentence, sentiment, subjective in rows:
        if sentiment == target_sent and subjective == target_subj: sentiment_subjective_sum += 1
    return sentiment_subjective_sum
calculateNumOfSS2(df.iloc[0,:]['AP'],2,1)
```

```python
sentiment_labels = ['negative', 'neutral', 'positive']
subjective_labels = ['subjective', 'objective']
```

```python
# neg/subj neg/obj neu/subj neu/obj pos/subj pos/obj
sentiment_subjective_list = []

for i in range(3):
    for j in range(2):
        ss = df['AP'].apply(lambda x: calculateNumOfSS2(x,i,j)).values
        sentiment_subjective_list.append(ss)
sentiment_subjective_list
```

```python
figure, axis = plt.subplots(3,2, figsize=(20, 10))

k = 0
for i in range(3):
    for j in range(2):
        label_name = f"{sentiment_labels[i]}/{subjective_labels[j]}"
        axis[i, j].plot(X,sentiment_subjective_list[k], label=label_name)
        axis[i, j].set_title(label_name)
        k += 1
plt.show()
```

```python
positive_list = df['AP'].apply(lambda x: calculateNumOfSS1(x,2,0)[0]).values
neutral_list = df['AP'].apply(lambda x: calculateNumOfSS1(x,1,0)[0]).values
negative_list = df['AP'].apply(lambda x: calculateNumOfSS1(x,0,0)[0]).values
negative_list
```

```python
plt.plot(X, negative_list, label='negative')
plt.plot(X, neutral_list, label='neutral')
plt.plot(X, positive_list, label='positive')
plt.legend()
plt.show()
```

```python
fig, axis = plt.subplots(3,1,figsize=(20,20))

axis[0].plot(X, negative_list)
axis[0].set_title('negative')
axis[1].plot(X, neutral_list)
axis[1].set_title('neutral')
axis[2].plot(X, positive_list)
axis[2].set_title('positive')

plt.show()
```

```python
subjective_list = df['AP'].apply(lambda x: calculateNumOfSS1(x,2,0)[1]).values
objective_list = df['AP'].apply(lambda x: calculateNumOfSS1(x,1,1)[1]).values
subjective_list
```

```python
plt.figure(figsize=(15,10))
plt.plot(X, subjective_list, label='subjective')
plt.plot(X, objective_list, label='objective')
plt.legend()
plt.show()
```
