```python
item_feature = pd.read_csv('item_feature.csv')
```

```python
item_feature.drop(columns='Unnamed: 0',inplace=True)
```

```python
import ast
for i in item_feature.count():
    ast.literal_eval(item_feature['feature'][0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer를 적용하기 위해 공백문자로 word 단위가 구분되는 문자열로 변환. 
item_feature['feature_literal'] = item_feature['feature'].apply(lambda x : (' ').join(x))
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(item_feature['feature_literal'])
print(genre_mat.shape)
```
