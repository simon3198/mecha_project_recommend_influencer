```python
import pandas as pd
```

```python
data = pd.read_csv('item_source.csv')
```

```python
data.fillna(0.0)
```

```python
item_list = data['Unnamed: 0'].tolist()
feature_list = data.columns.tolist()
feature_list = feature_list[1:]
```

```python
data.set_index('Unnamed: 0',inplace=True)
```

```python
item_dict={}
for item in item_list:
    item_dict[item]=[]
    for feature in feature_list:
        if float(data[feature][item]) == 1.0:
            item_dict[item].append(feature)

item_dict
```
